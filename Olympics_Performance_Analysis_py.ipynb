```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Create a mock dataset for India's Olympic performance
data = {
    "Year": [2000, 2004, 2008, 2012, 2016, 2020],  # Years of Olympics
    "Gold": [0, 1, 1, 0, 1, 1],
    "Silver": [0, 0, 0, 2, 1, 2],
    "Bronze": [1, 0, 2, 4, 1, 4],
    "Total_Medals": [1, 1, 3, 6, 3, 7],
    "Top_Sport": ["Weightlifting", "Shooting", "Boxing", "Shooting", "Badminton", "Wrestling"]
}

df = pd.DataFrame(data)
```

```python
# Step 2: Basic analysis
print("India's Olympic Performance Summary:")
print(df)

# Calculate medal percentage growth
df["Medal_Growth"] = df["Total_Medals"].pct_change() * 100
print("\nMedal Growth Percentage Over Years:")
print(df[["Year", "Total_Medals", "Medal_Growth"]])
```

```python

# Step 3: Visualizations

# Line chart for medals over the years
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Gold"], marker="o", label="Gold", color="gold")
plt.plot(df["Year"], df["Silver"], marker="o", label="Silver", color="silver")
plt.plot(df["Year"], df["Bronze"], marker="o", label="Bronze", color="brown")
plt.plot(df["Year"], df["Total_Medals"], marker="o", label="Total Medals", color="blue", linestyle="--")
plt.title("India's Olympic Medals Over the Years")
plt.xlabel("Year")
plt.ylabel("Medals")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar chart for Total Medals
plt.figure(figsize=(8, 6))
sns.barplot(x="Year", y="Total_Medals", data=df, palette="viridis")
plt.title("Total Medals by Year")
plt.ylabel("Total Medals")
plt.xlabel("Year")
plt.show()
```

```python
# Pie chart for medal distribution in 2020 Olympics
medals_2020 = df[df["Year"] == 2020][["Gold", "Silver", "Bronze"]].iloc[0]
medals_2020.plot.pie(
    autopct="%.1f%%", labels=["Gold", "Silver", "Bronze"], colors=["gold", "silver", "brown"],
    title="Medal Distribution (2020 Olympics)", figsize=(6, 6), startangle=90
)
plt.ylabel("")  # Remove default y-axis label
plt.show()
```

```python
# Step 4: Insights into top sports
top_sports = df["Top_Sport"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=top_sports.values, y=top_sports.index, palette="magma")
plt.title("India's Most Successful Sports Over the Years")
plt.xlabel("Count of Top Performance")
plt.ylabel("Sport")
plt.show()
```
